```python
import matplotlib.pyplot as plt
import numpy as np

# 
plt.style.use('seaborn-v0_8-whitegrid')

# 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# -
strain_ductile = np.array([0, 0.002, 0.004, 0.006, 0.008, 0.01, 0.02, 0.04, 0.06, 0.08, 0.1])
stress_ductile = np.array([0, 20, 40, 60, 80, 100, 100, 120, 110, 90, 80])

# -
ax1.plot(strain_ductile, stress_ductile, marker='o', color='blue')
ax1.set_title('Ductile Material', fontsize=12)
ax1.set_xlabel('Strain [-]', fontsize=10)
ax1.set_ylabel('Stress [ksi]', fontsize=10)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.annotate('Yield Point', xy=(0.01, 100), xytext=(0.015, 110),
             arrowprops=dict(facecolor='black', shrink=0.05))

# -
strain_brittle = np.array([0, 0.001, 0.002, 0.003, 0.004, 0.005])
stress_brittle = np.array([0, 50, 100, 150, 180, 190])

# -
ax2.plot(strain_brittle, stress_brittle, marker='o', color='red')
ax2.set_title('Brittle Material', fontsize=12)
ax2.set_xlabel('Strain [-]', fontsize=10)
ax2.set_ylabel('Stress [ksi]', fontsize=10)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.annotate('Fracture', xy=(0.005, 190), xytext=(0.006, 200),
             arrowprops=dict(facecolor='black', shrink=0.05))

# 
plt.tight_layout()

# 
plt.savefig('Stress_strain_curves.png', dpi=300)

# 
plt.show()
```
