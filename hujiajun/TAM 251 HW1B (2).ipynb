```python
import sympy as sy
import numpy as np
import math
from math import sin,cos,tan,asin,acos,atan,pi,sqrt,log,log2

g=9.81
cm=0.01
mm=0.001

def a2p(x):
    return x/180*pi

def theta_cos(a,b,c):
    return (a*a+b*b-c*c)/2/a/b

def c_cos(a,b,theta):
    return sqrt(a*a+b*b-cos(theta)*2*a*b)
```

知识重点
--- 
2.normal strain 正应变 $\epsilon=\frac{ΔL}{L_0}=\frac{长度变化}{原长}$  
3.normal stress 正应力 $\sigma=\frac{P}{A}$  
4.工程应变$=\frac{ΔL}{L_0}$ vs 实际应变$=\ln(\frac{L_1}{L_0})$  

```python
L = 220 # mm
a = 228 # mm
theta = 0.5 # degrees
alpha = 33 # degrees

theta=a2p(theta)
alpha=a2p(alpha)


##cos=(a*a+b*b-c*c)/2ab

BB=sqrt(a*a+a*a-cos(theta)*2*a*a)
cosB=(BB**2+a*a-a*a)/(2*BB*a)

angle_B=acos(cosB)
angle_A=alpha
angle_C=angle_A+angle_B

LL=c_cos(BB,L,angle_C)
(LL-L)/L*1000
```

```python
a = 11 # in
b = 6 # in
strain = 0.005

l0=sqrt(a*a+b*b)
l1=strain*l0+l0

#使用建系 向量法做题

x,y=sy.S('x,y')

#列B'A, B'D长度方程强解B'坐标
equ1=x**2+y**2-l1**2
equ2=(x-a)**2+y**2-b**2

sol=sy.solve([equ1,equ2],[x,y])
x=float(sol[0][0])
y=float(-sol[0][1])

#列向量
BB=np.array([x,y])
By=np.array([a,b])
Ax=np.array([a,0])

AB=By-Ax
ABB=BB-Ax

#点乘算cos角度
angle=sum(AB*ABB)/b/np.linalg.norm(ABB)
acos(angle)*180/pi
```
