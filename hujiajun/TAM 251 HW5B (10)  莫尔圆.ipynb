知识重点

莫尔圆 及其公式 
$$
\begin{aligned}
&\sigma_x'=\frac{\sigma_x+\sigma_y}{2}+\frac{\sigma_x-\sigma_y}{2}\cdot \cos(2\theta)+\tau_{xy}\sin(2\theta) \\ &\sigma_y'=\frac{\sigma_x+\sigma_y}{2}-\frac{\sigma_x-\sigma_y}{2}\cdot \cos(2\theta)-\tau_{xy}\sin(2\theta) \\  
&\tau_{xy}'=-\frac{\sigma_x-\sigma_y}{2}+\tau_{xy}\cos(2\theta) \\
\\
&C_{圆心}=\sigma_{平均}=\frac{\sigma_x+\sigma_y}{2}=\frac{\sigma_x'+\sigma_y'}{2}\\  
&R_{半径}=\sqrt{(\frac{\sigma_x-\sigma_y}{2})^2+\tau_{xy}^2}\\
\\
&\theta_p=\frac{1}{2}\tan^{-1}{(\frac{-\tau_{xy}}{\frac{\sigma_x-\sigma_y}{2}})}记得乘 \frac{1}{2}⭐⭐\\
\\
&\theta_s=\frac{1}{2}\tan^{-1}{(\frac{\frac{\sigma_x-\sigma_y}{2}}{-\tau_{xy}})}记得乘 \frac{1}{2}⭐⭐\\
\\
\end{aligned}
$$

**断裂标准**   
&nbsp;&nbsp;$FS=\frac{\sigma_Y}{\sigma_y}$  
1. **Von-mises 标准(对塑性材料)**   
$\sigma_{y}=\sqrt{\sigma_1^2-\sigma_1 \cdot \sigma_2+\sigma_2^2}$  
&nbsp;  
  
2. **Tresca 标准(对塑性材料)**   
$\sigma_{y}=\max{|\sigma_1|,|\sigma_2|,|\sigma_1-\sigma_2|}$  
&nbsp;  
  
3. **Maximum-normal 标准**

**Von-mises 标准**    
  
$
\begin{aligned}
\sigma_{y}&=\sqrt{\sigma_1^2-\sigma_1 \cdot \sigma_2+\sigma_2^2}\\
FS&=\frac{\sigma_Y}{\sigma_y}\\
\end{aligned}
$

```python
#计算sigma1
x=int(input('x='))
y=int(input('y='))
t=int(input('t='))

R=((x-y)**2/4+t**2)**0.5
C=0.5*x+0.5*y
s1=C+R
s2=C-R
ans=(s1**2-s1*s2+s2**2)**0.5
int((ans)*100)/100
```

**Tresca 标准**  
  
$
\begin{aligned}
\sigma_y&=\max{|\sigma_1|,|\sigma_2|,|\sigma_1-\sigma_2|}\\
FS&=\frac{\sigma_Y}{\sigma_y}\\
\end{aligned}
$

```python
#计算sigma1
x=int(input('x='))
y=int(input('y='))
t=int(input('t='))

R=((x-y)**2/4+t**2)**0.5
C=0.5*x+0.5*y
s1=(C+R)
s2=(C-R)
s3=(s1-s2)
int(max(abs(s1),abs(s2),abs(s3))*100)/100
```
