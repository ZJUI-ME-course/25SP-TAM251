```python
r=S('r')

b = 28 # mm
t = 8 # mm
Vy = 37 # kN

b*=mm
t*=mm
Vy*=1000

thick1=2*b
thick2=2*b
thick3=4*b+2*t

I1=integrate((4*b+2*t)*r*r,[r,-2*t,2*t])
I2=integrate(2*b*r*r,[r,2*t,2*t+2*b])*2
I3=integrate(4*b*r*r,[r,2*t+2*b,3*t+2*b])*2

I=I1+I2+I3


m1=t*(4*b)
m2=2*b*2*b
m3=2*t*(4*b+2*t)

y1=2*b+2.5*t
y2=2*t+b
y3=t



Q1=m1*y1
Q2=m1*y1+m2*y2
Q3=m1*y1+m2*y2+m3*y3

tau1=Vy*Q1/I/thick1
tau2=Vy*Q2/I/thick2
tau3=Vy*Q3/I/thick3

max(tau1/1000000,tau2/1000000,tau3/1000000)
```
