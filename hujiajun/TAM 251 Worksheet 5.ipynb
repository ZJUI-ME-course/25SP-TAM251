```python
from math import sin,cos,tan,pi,asin,acos,atan,sqrt
import sympy as sy
import numpy as np

w=30
h0=45
h1=37
h2=22

E=29*1000000
Y=36*1000
d=0.25
r=d/2
fs=2
ft=250

S=pi*r**2


s1=atan(w/h1)
s2=atan(w/h2)
```

```python
l1=sqrt(w**2+h1**2)
l2=sqrt(w**2+h2**2)
l3=l1
l4=l2

k1=E*S/l1
k2=E*S/l2
k3=k1
k4=k2

dd,fx,fy=sy.S('dd,fx,fy')
d1=h1*dd*sin(s1)
d2=h2*dd*sin(s2)
d3=d1
d4=d2

f1=k1*d1
f2=k2*d2
f3=k3*d3
f4=k4*d4

#equx=fx+ft-f1*sin(s1)-f2*sin(s2)+f3*sin(s1)+f4*sin(s2)
#equy=fy-f1*cos(s1)-f2*cos(s2)-f3*cos(s1)-f4*cos(s2)
#equm=-ft*h0+f1*sin(s1)*h1+f2*sin(s2)*h2-f3*sin(s1)*h1-f4*sin(s2)*h2

#sy.solve([equx,equy,equm],[dd,fx,fy])
f1/f2
```

```python
equm=-ft*h0+2*f1*sin(s1)*h1+2*f2*sin(s2)*h2
ans=sy.solve([equm],[dd])[dd]

f2=f2.subs(dd,ans) #lb
f1=f1.subs(dd,ans)

print(f1,f2)
```

```python
sigma1=f1/S/1000
sigma2=f2/S/1000
sigma_max=Y/fs/1000

print(sigma1,sigma2,sigma_max)
Y/sigma1
```
